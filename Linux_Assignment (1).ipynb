2. What is difference between hard link and soft link ?

 -> **Hard link**
    - Hard link is a mirror copy of the original file. It has no other names.
    - It contains the actual contents of the original file. Nothing happpens when the original file is removed.

   **Soft link**
   - Soft link is similar to the shortcut in windows. It can be reffered as symbolic link or symlink.
   - It contains the path of the original file and not the contents. Soft link when the original file is removed the link will be inaccessible because it points to a non-existent file. this is called hanging link.
